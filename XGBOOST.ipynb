```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import xgboost as xgb
```

```python
df = pd.read_csv(r'C:\Users\mvnik\OneDrive\Desktop\adult.csv')
df.head()
```

```python
df = df.drop(["education"], axis=1)
```

```python
df = pd.get_dummies(df)
df.head()
```

```python
df.columns
```

```python
df = df.drop("income_<=50K", axis = 1)
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train_full, X_test, y_train_full, y_test=train_test_split(X,y,random_state=0)
```

```python
X_train, X_val, y_train, y_val=train_test_split(X_train_full, y_train_full, random_state=0)
```

```python
from xgboost import XGBClassifier
xgb_clf = XGBClassifier(n_estimators = 500,
                        learning_rate = 0.05,
                        use_label_encoder = False,
                        eval_metric = "logloss",
                        early_stopping_rounds = 5,
                        n_jobs = -1)
```

```python
xgb_clf.fit(X_train, y_train,                    
            eval_set = [(X_val,y_val)],
            verbose = False)
```

```python
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
pred_test = xgb_clf.predict(X_test)
test_score = accuracy_score(pred_test, y_test)
print("Test score:", np.round(test_score,2))
```

```python
pred_train = xgb_clf.predict(X_train_full)
train_score = accuracy_score(pred_train, y_train_full)
print("Train score:", np.round(train_score,2))
```

```python
import folium
from geopy.geocoders import Nominatim
from folium import plugins
from folium.plugins import AntPath
from geographiclib.geodesic import Geodesic

# Initialize geolocator
geolocator = Nominatim(user_agent="geoapi")

# Get coordinates for Chennai and Kolkata
chennai = geolocator.geocode("Chennai, India")
kolkata = geolocator.geocode("Kolkata, India")

chennai_coords = (chennai.latitude, chennai.longitude)
kolkata_coords = (kolkata.latitude, kolkata.longitude)

# Create a folium map centered between Chennai & Kolkata
m = folium.Map(location=[20, 85], zoom_start=5)

# Add markers for both cities
folium.Marker(chennai_coords, popup="Chennai", icon=folium.Icon(color="blue")).add_to(m)
folium.Marker(kolkata_coords, popup="Kolkata", icon=folium.Icon(color="red")).add_to(m)

# Draw the straight-line route (Ship route approximation)
folium.PolyLine([chennai_coords, kolkata_coords], color="blue", weight=4, opacity=0.6).add_to(m)

# Optionally, add a curved geodesic line for a realistic ship route
geodesic_points = Geodesic.WGS84.InverseLine(chennai_coords[0], chennai_coords[1], kolkata_coords[0], kolkata_coords[1])
num_points = 50  # Increase for smoother curve

latitudes, longitudes = [], []
for i in range(num_points + 1):
    point = geodesic_points.Position(i / num_points * geodesic_points.s13)
    latitudes.append(point['lat2'])
    longitudes.append(point['lon2'])

folium.PolyLine(list(zip(latitudes, longitudes)), color="green", weight=3, opacity=0.7).add_to(m)

# Add AntPath animation (optional)
AntPath([chennai_coords, kolkata_coords], color="blue", delay=500).add_to(m)

# Save and display the map
m.save("ship_route_chennai_kolkata.html")
m
```
